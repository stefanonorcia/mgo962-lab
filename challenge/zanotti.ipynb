```python
import pandas as pd
from pandas import read_csv
from pandas import DataFrame
from pandas import Grouper
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('D:\\UNIVERSIT√Å\\TIME SERIES AND FORECASTING\\electricity.csv',index_col=0, parse_dates=True)
```

```python
groups = df.groupby(Grouper(freq='D'))
groupsw = df.groupby(Grouper(freq='W'))
```

```python
for name, group in groups:
    a = []
    b = []
    for i in range(0,len(group['Demand'])):
        a.append(str(group.index[i].hour)+ ':' + str(group.index[i].minute))
        b.append(group['Demand'][i])
    plt.plot(a,b)
    plt.xticks(rotation=90)
    plt.yticks(rotation=90)
plt.show()
```

```python
for name, group in groupsw:
    a = []
    b = []
    for i in range(0,len(group['Demand'])):
        a.append(str(group.index[i].dayofweek) + '-' + str(group.index[i].hour) + ':' + str(group.index[i].minute))
        b.append(group['Demand'][i])
    plt.plot(a,b)
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.show()
```
